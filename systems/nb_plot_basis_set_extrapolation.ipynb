```python
import numpy as np
import matplotlib.pyplot as pp
import pandas as pd
import os
import json

from functools import reduce
%matplotlib widget
```

```python
system = "./cu_nh3/nh3"
method = "standard"

folder = os.path.join(system, method)
metadata_file = os.path.join(folder, 'metadata.json')

with open(metadata_file, 'r') as inp:
    metadata = json.load(inp)
metadata
```

```python
datas = []
for basis in metadata['bases']:
    fname = os.path.join(folder, basis, f"table_nh3_0_{basis.lower()}.csv")
    data = pd.read_csv(
        fname, skipinitialspace=True,
        names=['GEM_BETA',basis], index_col=None, header=0
    )
    print(basis)
    
    datas.append(data)
display(data)
df_merged=reduce(
    lambda left, right: pd.merge(
        left, right, on='GEM_BETA'
    ), 
    datas)

df_merged
```

```python
df_merged.plot(x = 'GEM_BETA')
```
